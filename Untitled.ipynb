```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
!whereis python
```

## Load  dataset

```python
train = pd.read_csv("archive/Training.csv")
test = pd.read_csv("archive/Testing.csv")
```

```python
symptoms = open("symptoms","w")
symptoms.write("[\"")
for x in train.columns:
    symptoms.write(str(x)+"\",\"")
symptoms.write("]")
symptoms.close()
```

```python
train = train.drop(["Unnamed: 133"],axis=1)
print(train.columns)
```

```python
# train[train['itching']==1]
plt.bar(["Itching","No itching"],[len(train[train["itching"]==1]),len(train[train["itching"]!=1])])
```

```python
# train[train['itching']==1]
for x in train.columns:
    plt.bar([x,"No "+str(x)],[len(train[train[x]==1]),len(train[train[x]!=1])])
    plt.show()
```

```python
train_itching = train[train["itching"]==1]
# print(train_itching)
print(len(train_itching))
x=[]
y=[]
for temp in train_itching["prognosis"].unique():
    x.append(temp)
    y.append(len(train_itching[train_itching["prognosis"]==temp]))
print(x,y)
plt.pie(y,labels=x)
```

```python
# train_itching = train[train["itching"]==1]
# print(train_itching)
# print(len(train_itching))
x=[]
y=[]
for temp in train["prognosis"].unique():
    x.append(temp)
    y.append(len(train[train["prognosis"]==temp]))
# print(x,y)
plt.pie(y,labels=x)
plt.show()
# plt.xaxis.set_tick_params(pad = 5)
# man.show()
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
y_train=train["prognosis"].values
x_train=train.drop(["prognosis"], axis=1)
# print(x,y)
nb_model = nb_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
y_test=test["prognosis"].values
x_test=test.drop(["prognosis"], axis=1)
# print(x,y)
y_pred = nb_model.predict(x_test)
print(classification_report(y_test, y_pred))
```

## Saving the model

```python
import pickle as pkl
with open("NaiveBayesModel.sav","wb") as model:
    pkl.dump(nb_model,model)
```
